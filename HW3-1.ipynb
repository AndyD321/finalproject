```python
features = df.columns[:5948] 
X = df[features].values
labels = df.columns[-1]
y = df[labels].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
tree = DecisionTreeClassifier() # decision tree algorithm
tree.fit(X_train, y_train)
```

```python
# a) Decision Tree (Optimization)
param_grid = { 'max_depth': [2, 3, 4, 5, 6, 7], 'min_samples_leaf': [2, 20, 50,70], 
               'max_leaf_nodes': [5, 9, 15, 21, 27]}
gs_dt = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5, n_jobs=-1)
gs_dt.fit(X_train, y_train)
```

```python
print(f'The best model is: \n{gs_dt.best_estimator_}')
tree_opt = gs_dt.best_estimator_
print(f'Decision tree has {tree_opt.tree_.node_count} nodes with maximum depth {tree_opt.tree_.max_depth}.') 
print(f'Model Accuracy: {tree_opt.score(X_train, y_train)}')
y_pred = tree_opt.predict(X_test) 
print(classification_report(y_test,y_pred))
```

```python
from sklearn.utils import resample

df_class1 = resample(df[df['Classes']=='Class1'],
               replace=True,
               n_samples=118,
               random_state=123)

df_class2 = resample(df[df['Classes']=='Class2'],
               replace=True,
               n_samples=118,
               random_state=123)

df_class3 = resample(df[df['Classes']=='Class3'],
               replace=True,
               n_samples=118,
               random_state=123)

df_class4 = resample(df[df['Classes']=='Class4'],
               replace=True,
               n_samples=118,
               random_state=123)

df_class5 = resample(df[df['Classes']=='Class5'],
               replace=True,
               n_samples=118,
               random_state=123)

df_upsampled = pd.concat([df_class1, df_class2, df_class3, df_class4, df_class5])

df_upsampled.Classes.value_counts()
```

```python
features = df_upsampled.columns[:5948] 
X = df_upsampled[features].values
labels = df.columns[-1]
y = df_upsampled[labels].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
```

```python
param_grid = { 'max_depth': [2, 3, 4, 5, 6, 7], 'min_samples_leaf': [2, 20, 50,70], 
               'max_leaf_nodes': [5, 9, 15, 21, 27]}
gs_dt = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5, n_jobs=-1)
gs_dt.fit(X_train, y_train)
```

```python
print(f'The best model is: \n{gs_dt.best_estimator_}')
tree_opt = gs_dt.best_estimator_
print(f'Decision tree has {tree_opt.tree_.node_count} nodes with maximum depth {tree_opt.tree_.max_depth}.') 
print(f'Model Accuracy: {tree_opt.score(X_train, y_train)}')
y_pred = tree_opt.predict(X_test) 
print(classification_report(y_test,y_pred))
```
